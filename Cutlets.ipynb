```python
#find variance for each group
print(np.var(UnitA), np.var(UnitB))
```

```python
import scipy.stats as stats
stats.ttest_ind(a=UnitA, b=UnitB, equal_var=True)
```

```python
 # because the p value is greater than the alpha value (0.05), we failed to reject the null hypothesis. we do not have sufficient evidence to prove that th cutleries have a different diameter size
```
